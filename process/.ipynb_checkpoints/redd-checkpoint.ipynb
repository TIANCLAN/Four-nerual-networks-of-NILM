```python
for appliance_name in name:
    #print(appliance_name[0]) #housex的x电器
    data = next(elec[appliance_name[1]].load()) #提取h5文件中存储的真实参数名（无空格）

    #整理格式
    data['active'] = data['power']['active']
    data['time'] = data.index
    data['time'] = pd.to_numeric(data['time'])
    #分别将单个电器数据数值化后的日期以及实功率写入npy文件中
    file=[]
    for row in data.itertuples():
        file.append([float(getattr(row, '_3')),float(getattr(row, '_2'))])
    print(file[:20])
    #写入house_x文件夹
    np.save(join(folder+'/house_'+str(building), '{}_appliance.npy'.format(appliance_name[0])), file)
    data=[]
print('finished!')
```

```python
#时间序列main与单个用电器对齐，算法真正需要的数据
base_path = r'../../Data/REDD/'
for appliance_name in name:
        for house_id in [1,2,5]:
            aggregate_new = []
            appliance_new = []

            aggregate_data = np.load(join(base_path, 'house_%d/main.npy' % house_id))
            appliance_data = np.load(join(base_path, 'house_%d/%s_appliance.npy' % (house_id, appliance_name[0])))

            aggregate_index = 0
            appliance_index = 0
            if appliance_data[0][0] < aggregate_data[0][0]:
                print('Appliance time is ahead of aggregate time!')
                for i in range(0, len(appliance_data)):
                    if appliance_data[i][0] < aggregate_data[0][0]:
                        appliance_index = i

            for i in range(appliance_index, len(appliance_data)):
                appliance_time = appliance_data[i][0]
                front = aggregate_index
                behind = -1
                for j in range(aggregate_index, len(aggregate_data)):
                    if aggregate_data[j][0] <= appliance_time:
                        front = j
                    else:
                        behind = j
                        d_1 = abs(appliance_time - aggregate_data[front][0])
                        d_2 = abs(appliance_time - aggregate_data[behind][0])
                        if d_1 <= d_2:
                            aggregate_new.append(aggregate_data[front][1])
                            appliance_new.append(appliance_data[i][1])
                        else:
                            aggregate_new.append(aggregate_data[behind][1])
                            appliance_new.append(appliance_data[i][1])
                        aggregate_index = front
                        break
                if i % 10000 == 0:
                    print('Appliance: %s, House: %s, processing: %f' % (
                    appliance_name, house_id, i / len(appliance_data)))
            print("Appliance: {}, House: {} finished!".format(appliance_name[0], house_id))
            np.save(join(base_path, "{}_main_house_{}.npy".format(appliance_name[0], house_id)), aggregate_new)
            np.save(join(base_path, "{}_appliance_house_{}.npy".format(appliance_name[0], house_id)), appliance_new)
```
