```python
CMRatio = [0,0,0,0]
DWRatio = [0,0,0,0]
FridgeFreezerRatio = [0,0,0,0]
MicrowaveRatio = [0,0,0,0]
televisionRatio = [0,0,0,0]
toasterRatio = [0,0,0,0]
for file in files:
    test=np.load(file, encoding = "latin1",allow_pickle=True) #加载文件
    explainRatio = max(test[1])
    if("ComputerMonitor" in file):
        if("VAE" in file):
            CMRatio[0]=explainRatio
        if("DAE" in file):
            CMRatio[1]=explainRatio
        if("S2S" in file):
            CMRatio[2]=explainRatio
        if("S2P" in file):
            CMRatio[3]=explainRatio
    if("Dishwasher" in file):
        if("VAE" in file):
            DWRatio[0]=explainRatio
        if("DAE" in file):
            DWRatio[1]=explainRatio
        if("S2S" in file):
            DWRatio[2]=explainRatio
        if("S2P" in file):
            DWRatio[3]=explainRatio
    if("FridgeFreezer" in file):
        if("VAE" in file):
            FridgeFreezerRatio[0]=explainRatio
        if("DAE" in file):
            FridgeFreezerRatio[1]=explainRatio
        if("S2S" in file):
            FridgeFreezerRatio[2]=explainRatio
        if("S2P" in file):
            FridgeFreezerRatio[3]=explainRatio
    if("Microwave" in file):
        if("VAE" in file):
            MicrowaveRatio[0]=explainRatio
        if("DAE" in file):
            MicrowaveRatio[1]=explainRatio
        if("S2S" in file):
            MicrowaveRatio[2]=explainRatio
        if("S2P" in file):
            MicrowaveRatio[3]=explainRatio
    if("television" in file):
        if("VAE" in file):
            televisionRatio[0]=explainRatio
        if("DAE" in file):
            televisionRatio[1]=explainRatio
        if("S2S" in file):
            televisionRatio[2]=explainRatio
        if("S2P" in file):
            televisionRatio[3]=explainRatio
    if("toaster" in file):
        if("VAE" in file):
            toasterRatio[0]=explainRatio
        if("DAE" in file):
            toasterRatio[1]=explainRatio
        if("S2S" in file):
            toasterRatio[2]=explainRatio
        if("S2P" in file):
            toasterRatio[3]=explainRatio
    print(file,':',explainRatio)

print(CMRatio)
print(FridgeFreezerRatio)
print(televisionRatio)
print(toasterRatio)
print(MicrowaveRatio)
print(DWRatio)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def Draw_Parallel_bar(features,L1,L2,L3,L4,L5,L6):
    features=features
    RA=L1;SS=L2;SLE=L3;Arthritis=L4;
    plt.figure(figsize=(7,4))
    bar_width = 0.1
    index_RA = np.arange(len(features))
    index_SS = index_RA + bar_width
    index_SLE = index_SS + bar_width
    index_Arthritis = index_SLE + bar_width
    index_L5 = index_Arthritis + bar_width
    index_L6 = index_L5 + bar_width

    font = {'family': 'Times New Roman',
            'size': 12,
            }
    sns.set(font_scale=1.2)
    plt.rc('font',family='Times New Roman')

    plt.bar(index_RA, height=RA, width=bar_width, color='peru', label='computerMonitor')
    plt.bar(index_SS, height=SS, width=bar_width, color='g', label='Dishwasher')
    plt.bar(index_SLE, height=SLE, width=bar_width, color='y', label='Freezer')
    plt.bar(index_Arthritis, height=Arthritis, width=bar_width, color='c', label='Microwave')
    plt.bar(index_L5, height=L5, width=bar_width, color='red', label='television')
    plt.bar(index_L6, height=L6, width=bar_width, color='grey', label='toaster')

    plt.legend(fontsize = 12,bbox_to_anchor=(1.0,0),loc=3)  # 显示图例
    plt.xticks(index_RA + 10*bar_width/5, features,fontsize = 10)
    plt.xlabel('Building--house3',fontsize = 14)
    plt.ylabel('Total energy explaned',fontsize = 14)
    
    plt.savefig('house3',bbox_inches = 'tight')
    plt.show()
    

if __name__=="__main__":
    print(CMRatio)
    print(DWRatio)
    print(FridgeFreezerRatio)
    print(MicrowaveRatio)
    print(televisionRatio)
    print(toasterRatio)
    #Draw_Parallel_bar(('VAE', 'DAE', 'S2S', 'S2P'),
    #                  CMRatio,DWRatio,FridgeFreezerRatio,MicrowaveRatio,televisionRatio,toasterRatio)
    
```
